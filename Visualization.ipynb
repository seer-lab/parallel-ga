```python
def plotModel(test, test1, test2, test3, name):
    plt.figure(figsize=(20,8))
    plt.xlabel("Runs")
    plt.ylabel("Time (seconds)")
    plt.title(name)

    plt.plot(test['Real Time'].iloc[0:10], 'r.-', label="Hybrid Model", linewidth=2.5)
    plt.plot(test1['Real Time'].iloc[0:10], 'b.-', alpha=.5, label="Island Model")
    plt.plot(test2['Real Time'].iloc[0:10], 'g.-', label="Main Secondary Model")
    plt.plot(test3['Real Time'].iloc[0:10], 'y.-', alpha=.5, label="Serial Model")
    plt.legend(loc="center right")
```

```python
def plot4Models(test, test1, test2, test3, name, test4, test5, test6, test7, name1, 
                test8, test9, test10, test11, name2, test12, test13, test14, test15, name3):
    f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
    f.set_figwidth(20)
    f.set_figheight(10)
    
    ax1.set_title(name)
    ax1.plot(test['Real Time'].iloc[0:10], 'r.-', label="Hybrid Model", linewidth=2.5)
    ax1.plot(test1['Real Time'].iloc[0:10], 'b.-', alpha=.5, label="Island Model")
    ax1.plot(test2['Real Time'].iloc[0:10], 'g.-', label="Main Secondary Model")
    ax1.plot(test3['Real Time'].iloc[0:10], 'y.-', alpha=.5, label="Serial Model")
    ax1.set_xlabel("Runs")
    ax1.set_ylabel("Time (seconds)")
    ax1.legend(loc="center right")
    
    ax2.set_title(name1)
    ax2.plot(test4['Real Time'].iloc[0:10], 'r.-', label="Hybrid Model", linewidth=2.5)
    ax2.plot(test5['Real Time'].iloc[0:10], 'b.-', alpha=.5, label="Island Model")
    ax2.plot(test6['Real Time'].iloc[0:10], 'g.-', label="Main Secondary Model")
    ax2.plot(test7['Real Time'].iloc[0:10], 'y.-', alpha=.5, label="Serial Model")
    ax2.set_xlabel("Runs")
    ax2.set_ylabel("Time (seconds)")
    ax2.legend(loc="center right")
    
    ax3.set_title(name2)
    ax3.plot(test8['Real Time'].iloc[0:10], 'r.-', label="Hybrid Model", linewidth=2.5)
    ax3.plot(test9['Real Time'].iloc[0:10], 'b.-', alpha=.5, label="Island Model")
    ax3.plot(test10['Real Time'].iloc[0:10], 'g.-', label="Main Secondary Model")
    ax3.plot(test11['Real Time'].iloc[0:10], 'y.-', alpha=.5, label="Serial Model")
    ax3.set_xlabel("Runs")
    ax3.set_ylabel("Time (seconds)")
    ax3.legend(loc="center right")
    
    ax4.set_title(name3)
    ax4.plot(test12['Real Time'].iloc[0:10], 'r.-', label="Hybrid Model", linewidth=2.5)
    ax4.plot(test13['Real Time'].iloc[0:10], 'b.-', alpha=.5, label="Island Model")
    ax4.plot(test14['Real Time'].iloc[0:10], 'g.-', label="Main Secondary Model")
    ax4.plot(test15['Real Time'].iloc[0:10], 'y.-', alpha=.5, label="Serial Model")
    ax4.set_xlabel("Runs")
    ax4.set_ylabel("Time (seconds)")
    ax4.legend(loc="center right")
    
    plt.show()
```

```python
plot4Models(Hybrid_Model_1_1, GA_Island_Global_Mem_1_1, Main_Secondary_GA_1_1, Serial_GA_1_1, "Sphere Eval-Arithmetic Crossover",
           Hybrid_Model_2_1, GA_Island_Global_Mem_2_1, Main_Secondary_GA_2_1, Serial_GA_2_1, "Rastrigin Eval-Arithmetic Crossover",
           Hybrid_Model_3_1, GA_Island_Global_Mem_3_1, Main_Secondary_GA_3_1, Serial_GA_3_1, "Ackley Eval-Arithmetic Crossover",
           Hybrid_Model_4_1, GA_Island_Global_Mem_4_1, Main_Secondary_GA_4_1, Serial_GA_4_1, "Griewangk Eval-Arithmetic Crossover")
```
